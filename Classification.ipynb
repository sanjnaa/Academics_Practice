```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
#classification using logistic regresion
#sigma function for two values zero and postitve(1)
df = load_iris()
```

```python
df
```

```python
d = pd.DataFrame(df['data'],columns =df.feature_names)
```

```python
d
```

```python
d=d.head(100)
d
```

```python
df2 = pd.DataFrame(df['target'],columns =['Species'])
```

```python
df2
```

```python
df2=df2.head(100)
```

```python
df2
```

```python
plt.scatter(d['sepal length (cm)'],d['sepal width (cm)'],color='g')
```

```python
plt.scatter(d['petal length (cm)'],d['petal width (cm)'],marker="*",color="r")
```

```python
plt.plot(d)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(d,df2,test_size = 0.3)
```

```python
reg.fit(x_train,y_train)
```

```python
a=reg.predict(x_test)
a
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,a)
```
