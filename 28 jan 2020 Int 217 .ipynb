```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron,LogisticRegression
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score
```

```python
data = load_iris()
```

```python
x = data.data
```

```python
y = data.target
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
X = x[:,0:2]
```

```python
for i in range(0,4):
    for j in range(i+1,4):
        plt.scatter(x[:,i],x[:,j],c=y)
        plt.show()
```

```python
for i in range(0,4):
    for j in range(i+1,4):
        X = x[:,[i,j]]
        xtrain,xtest,ytrain,ytest = tts(X,y,test_size = 0.2,random_state = 0)
        clf = Perceptron(eta0=0.5,max_iter=1000,tol=1e-3,penalty='l2')
        clf.fit(xtrain,ytrain)
        pred = clf.predict(xtest)
        print("accuracy",i ,j ,"=",accuracy_score(ytest,pred))
        
        
```

```python
for i in range(0,4):
    for j in range(i+1,4):
        X = x[:,[i,j]]
        y = y
        xtrain,xtest,ytrain,ytest = tts(X,y,test_size = 0.2,random_state = 0)
        clf1 = LogisticRegression(solver='lbfgs',max_iter=100,multi_class='auto') 
        clf1.fit(xtrain,ytrain)
        print("train_accuracy",clf1.score(xtrain,ytrain))
        pred = clf1.predict(xtest)
        print("accuracy",i ,j ,"=",accuracy_score(ytest,pred))
        
        
```
