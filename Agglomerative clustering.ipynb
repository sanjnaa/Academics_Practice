```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as p
```

```python
np.random.seed(123)
```

```python
variables=['x','y','z']

#labels=['ID1','ID2','ID3','ID4','ID5']
X=np.random.random([50,3])*10
```

```python
df=pd.DataFrame(X,columns=variables)
df
#,index=labels
```

```python
from sklearn.cluster import AgglomerativeClustering as ac
```

```python
a=ac(n_clusters=3,affinity='euclidean',linkage='complete')
labels=a.fit_predict(X)
```

```python
'Cluster label:%s'%labels
```

```python
p.scatter(X[labels==0,0],X[labels==0,1],c='red')
p.scatter(X[labels==1,0],X[labels==1,1],c='green')
p.scatter(X[labels==2,0],X[labels==2,1],c='yellow')
p.show()
```

```python
#dendogram
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage,dendrogram
row_clusters=linkage(df.values,method='complete',metric='euclidean')
df1=pd.DataFrame(row_clusters,columns=['rowlabel1','rowlabel2','distance','no. of items in cluster'],index=['cluster %d'%(i+1) for i in range(row_clusters.shape[0])])
df1
```

```python
row_dendr=dendrogram(row_clusters,labels=labels)
plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()
```
