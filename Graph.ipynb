```python
#graph of data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x =  np.linspace(0,5,11)
x
```

```python
y=x**2
```

```python
#plotting and its attributes
plt.plot(x,y,'r')# graph line color is set by usin g r
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Title")
```

```python
plt.show()
```

```python
#subplot
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])#main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])#inset axes

#larger figure axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('xlabel_axes1')
axes1.set_ylabel('ylabel_axes1')
axes1.set_title('Axes 1 title')

# Insert Figure Axes 2
axes2.plot(x,y,'r')
axes2.set_xlabel('xlabel_axes2')
axes2.set_ylabel('ylabel_axes2')
axes2.set_title('Axes 2 title')
```

```python
fig, axes=plt.subplots(figsize=(12,3))

axes.plot(x,y,'r')
axes.set_xlabel('xlabel_axes')
axes.set_ylabel('ylabel_axes')
axes.set_title('Axes  title')
```

```python
fig.savefig("graph_Img.png")
```

```python
#legends,labels and titles
#will show which line represent
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend()

```

```python
plt.scatter(x,y,marker="*",color="g")
```
