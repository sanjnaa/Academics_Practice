```python
#Decision tree
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
df = load_iris()
```

```python
df
```

```python
d = pd.DataFrame(df['data'],columns =df.feature_names)
```

```python

d
```

```python
df2 = pd.DataFrame(df['target'],columns =['Species'])
```

```python

df2
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(d,df2,test_size = 0.3)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
DecTree=DecisionTreeClassifier()
```

```python
a = DecTree.fit(x_train,y_train)
```

```python
a
```

```python
pred=DecTree.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
from IPython.display import Image
from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz
import pydot


features = ['sepal length(cm)','sepal width(cm)','petal length(cm)','petal width(cm)']
features


dot_data = StringIO()
export_graphviz(DecTree, out_file = dot_data, feature_names = features, filled = True, rounded = True, special_characters = True)
graph = pydot.graph_from_dot_data(dot_data.getvalue())
Image(graph[0].create_png())
```
