```python
from sklearn.datasets import make_moons as mm
X,y=mm(n_samples=200,noise=0.05,random_state=0)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans as km
from sklearn.cluster import AgglomerativeClustering as ac
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
k=km(n_clusters=2,random_state=0)
ykm=k.fit_predict(X)
plt.scatter(X[ykm==0,0],X[ykm==0,1],c='lightblue',edgecolor='black',marker='o',s=40,label='cluster1')
plt.scatter(X[ykm==1,0],X[ykm==1,1],c='red',edgecolor='black',marker='s',s=40,label='cluster2')
plt.title('K-means Clustering')
a=ac(n_clusters=2,affinity='euclidean',linkage='complete')
yac=a.fit_predict(X)
plt.subplot(1,2,2)
plt.scatter(X[yac==0,0],X[yac==0,1],c='lightblue',edgecolor='black',marker='o',s=40,label='cluster1')
plt.scatter(X[yac==1,0],X[yac==1,1],c='red',edgecolor='black',marker='s',s=40,label='cluster2')
plt.title('Agglomerative Clustering')
plt.legend()
```

```python
from sklearn.cluster import DBSCAN as db
```

```python
d=db(eps=0.2,min_samples=5,metric='euclidean')
ydb=d.fit_predict(X)
plt.figure()
plt.scatter(X[ydb==0,0],X[ydb==0,1],c='lightblue',edgecolor='black',marker='o',s=40,label='cluster1')
plt.scatter(X[ydb==1,0],X[ydb==1,1],c='red',edgecolor='black',marker='s',s=40,label='cluster2')
plt.title('DBSCAN Clustering')
plt.legend()
plt.show()
```
