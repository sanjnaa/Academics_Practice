```python
df1=pd.DataFrame({'A':['AB','A1','A2','A3'],
                 'B':['B0','B1','B2','B3'],
                 'C':['Cb','Cv','Cx','Cz'],
                 'D':['Dx','Dv','Db','Dd']},
                index=[0,1,2,3])

df2=pd.DataFrame({'A':['AB','A1','A2','A3'],
                 'B':['zB0','zB1','zB2','zB3'],
                 'C':['Czb','Czv','Czx','Czz'],
                 'D':['Dbx','Dbv','Dbb','Dbd']},
                index=[0,1,2,3])

df3=pd.DataFrame({'A':['AB','A1','A2','A3'],
                 'B':['B11','B12','B22','B23'],
                 'C':['C2b','C2v','C2x','C2z'],
                 'D':['D1x','D1v','D1b','D1d']},
                index=[0,1,2,3])
```

```python
left=pd.DataFrame({'A':['AB','A1','A2','A3'],
                 'B':['B0','B1','B2','B3'],              
                 },
                index=['k0','k1','k2','k3'])
right=pd.DataFrame({'C':['Cb','Cv','Cx','Cz'],
                 'D':['Dx','Dv','Db','Dd']},
                index=['k0','k2','k3','k4'])
```

```python
left.join(right)
```

```python
left.join(right,how='outer')
```

```python
left.join(right,how='inner')
```
