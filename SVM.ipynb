```python
from sklearn import svm
clf =  svm.SVC(gamma = 0.001, C=100)#svr for regression and c for classsification
```

```python
x,y =  digit.data[:-1],digit.target[:-1]
```

```python
clf.fit(x,y)
```

```python
print("prediction :",clf.predict(digit.data[[-1]]))
```
